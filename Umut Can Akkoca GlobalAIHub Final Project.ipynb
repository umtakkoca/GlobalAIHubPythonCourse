```python
class Question:
    def __init__(self, prompt, answer):
        self.prompt = prompt
        self.answer = answer
            
print("SPOR QUIZİ KURALLARI:")
print("1.Her sorunun bir cevabı vardır.")
print("2.Boş bırakma hakkınız vardır.")
print("3.Büyük-küçük harf farketmeksizin şıkkınızı girebilirsiniz.")
print("BAŞARILAR :)\n")


question_prompts = [
     "1. Hangi futbolcu Fenerbahçe forması giymiştir?\n(A)Arda Turan\n(B)Sergen Yalçın\n(C)Selçuk İnan\n(D)Cenk Tosun\n",
     "2. Aşağıdaki takımlardan hangisi Türkiye Süper Liginde şampiyon olmuştur??\n(A)Bursaspor\n(B)Ankaragücü\n(C)Eskişehirspor\n(D)Samsunspor\n",
     "3. Galatasaray hangi yıl UEFA kupasını almıştır?\n(A)1999\n(B)2000\n(C)2001\n(D)2002\n",
     "4. Bir dönem Pelé, Franz Beckenbauer, Carlos Alberto ve Yasin Özdenak ‘ın da oynadığı ABD futbol takımı hangisidir?\n(A)Los Angeles Galaxy\n(B)DC United\n(C)Chicago Fire\n(D)New York Cosmos\n",
     "5. 1930 yılında oynanan ilk Dünya Kupası'nı hangi ülke kazanmıştır?\n(A)Arjantin\n(B)Brezilya\n(C)Uruguay\n(D)Almanya\n",
     "6. Diego Maradona hangi maçta “Tanrı’nın eli” diye adlandırılan golü atmıştır?\n(A)Arjantin-Türkiye\n(B)Arjantin-Fransa\n(C)Arjantin-İngiltere\n(D)Arjantin-İtalya\n",
     "7. Avrupa'da maç başına ortalama taraftar sayısı en fazla olan kulüp hangisidir?\n(A)Borussia Dortmund\n(B)Bayern Munchen\n(C)Real Madrid\n(D)Manchester United\n",
     "8. Dünya Kupası tarihinde en çok gol atan futbolcu kimdir?\n(A)Pele\n(B)Maradona\n(C)Ronaldo\n(D)Miroslav Klose\n",
     "9. İstanbul'un ev sahipliği yaptığı unutulmaz Liverpool-Milan Şampiyonlar Ligi finali hangi sene oynanmıştır?\n(A)2004\n(B)2005\n(C)2006\n(D)2007\n",
     "10. Millî takım formasını en fazla giyen futbolcu kimdir?\n(A)Rüştü Reçber\n(B)Hakan Şükür\n(C)Emre Belözoğlu\n(D)Arda Turan\n",
     
]

questions = [
     Question(question_prompts[0], "B"),
     Question(question_prompts[1], "A"),
     Question(question_prompts[2], "B"),
     Question(question_prompts[3], "D"),
     Question(question_prompts[4], "C"),
     Question(question_prompts[5], "C"),
     Question(question_prompts[6], "A"),
     Question(question_prompts[7], "D"),
     Question(question_prompts[8], "B"),
     Question(question_prompts[9], "A"),
]

def run_quiz(questions):
    score = 0
    empty = 0
        
    for question in questions:
        answer = input(question.prompt)
        answer = answer.upper()
          
        
        if answer == "":
            empty += 1
               
          
        elif answer == question.answer:
                  score += 10
                    
    if empty >= 5:
        print("5 soru veya daha az soru cevapladınız. BAŞARISIZ!!!")
        print("Az soru cevapladığınız için cevap anahtarını göremeyeceksiniz. Tekrar Deneyin!")
    
    elif empty < 5 and score < 51:
        print("100 üzerinden", score, "aldınız. BAŞARISIZ!!!")
        print("Cevap Anahtarı: 1.B 2.A 3.B 4.D 5.C 6.C 7.A 8.D 9.B 10.A")
    
    elif empty < 5 and score > 51:
        print("100 üzerinden", score, "aldınız. BAŞARILI!!!")
        print("Cevap Anahtarı: 1.B 2.A 3.B 4.D 5.C 6.C 7.A 8.D 9.B 10.A")
              
    

run_quiz(questions)
```
